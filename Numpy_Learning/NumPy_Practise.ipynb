<h3> Copies and Views</h3>

A slicing operation creates a view on original array, which is just a way of accessing array data. thus the original array is not copied in memory. you can  use np.may_share_memory() to check if two arrays share the same memory blockk

```python
#When modifying the view, the original array is modified as well
a=np.arange(10)
a
```

```python
b=a[::2]  # a and b both are pointing to same array 0,1,2..9 for memory optimization but b is actually sub array. so here copy of aray is not used.
b
```

```python
np.shares_memory(a,b)
```

```python
b[0]=10
b
```

```python
# eventhough we modified b, it updates 'a' because both shares same memory
a
```

```python
a=np.arange(10)
c=a[::2].copy()  # Force a copy
c
```

```python
np.shares_memory(a,c)  # we forcelly say not to share memory.. a and c are pointing other locations. Now if u modified 'c'; then 'a' will not get updated
```

```python
c[0]=10
c
```

```python
a  # a is not updating
```

<h3> Fancy Indexing </h3>

Numpy Array can be indexed with slices, but also with boolean or integer arrays(masks). this method is called fancy indexing. It creates copies not views.

```python
# using boolean mask
a=np.random.randint(0,20,15)  # generates random number between 0 to 20 and generates 15 such numbers which all are of integer
print(a.dtype)
a
```

```python
mask=(a%2==0)
mask
```

```python
extract_from_a=a[mask]  # where the a%2==0 become true ; those values come in extract_from_a
extract_from_a
```

```python
# indexing with a mask can be very useful to assign a new value to a sub array
a[mask]=-1
a # all even numbers repleced by -1
```

```python
# indexing with an array of integeres
a=np.arange(0,100,10)
a
```

```python
a[[2,3,2,4,2]] # here we are giving indices in the list like 2nd index,3rd index, again 2nd index and so on and print their values
```

```python
a[[9,7]]=-200  # on index 9 and 7 .. replace with -200
a
```
