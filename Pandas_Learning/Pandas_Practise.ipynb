<h2> Group By</h2>

```python
df=pd.read_csv('weather_data_cities.csv')
df  #weather by cities
```

```python
g=df.groupby('city')   # it will form the groups based on city
print(g)
```

```python
# we are printing city name along with its whole dataframe
for city, city_df in g:
  print(city)
  print('--------')
  print(city_df)
  print('***************************************************************')
#indices will be copied from original data so in first city mumbai....indices start from 4 5 6 7..
```

```python
# or to get specific group
g.get_group('new york')
```

```python
# Find maximum temerature and maximum windspeed in each of the cities
print(g.max())
```

```python
print(g.mean())
```

```python
print(g.describe())  # describe will run on only numerical type column
```

<h2>Concatenate Data Frames</h2>

```python
#make one df using dictionary or also called hash map
india_weather=pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temeperature":[32,45,30],
    "humidity":[80,60,78]
})
india_weather
```

```python
guj_weather=pd.DataFrame({
    "city":["ahmedabad","idar","surat"],
    "temeperature":[38,45,32],
    "humidity":[70,63,75]
})
guj_weather
```

```python
# Concate two dataframes
df=pd.concat([india_weather,guj_weather])
df
```

```python
# if we want continuous index
df=pd.concat([india_weather,guj_weather],ignore_index=True)
df
```

```python
df=pd.concat([india_weather,guj_weather],axis=1)  # we want to concate column wise not row wise
df
```

<h2> Merge Data Frames
</h2>

```python
temperature_df=pd.DataFrame({
    "city":["mumbai","delhi","banglore","hyderabad"],
    "temperature":[32,45,30,40]
})
temperature_df
```

```python
humidity_df=pd.DataFrame({
    "city":["delhi","mumbai","banglore"],
    "humidity":[68,65,75]
})
humidity_df
```

```python
# merge two dataframes without explicitly mention index
df=pd.merge(temperature_df,humidity_df,on='city')
df
```

```python
# Outer Join
df=pd.merge(temperature_df,humidity_df,on='city',how='outer')  # same like dbms
df
```

<h2> Numerical Indexing(.loc vs iloc)
</h2>

```python
df=pd.DataFrame([1,2,3,4,5,6,7,8,9,19],index=[49,48,47,46,45,1,2,3,4,5]) # creates indexing..row number is default ndex...here we r not using default indexing
df
```

```python
df=pd.DataFrame([1,2,3,4,5,6,7,8,9,19])  # default indexing
df
```

```python
df=pd.DataFrame([1,2,3,4,5,6,7,8,9,19],index=[49,48,47,46,45,1,2,3,4,5]) # creates indexing..row number is default ndex...here we r not using default indexing
df
```

```python
# access elements
df.loc[:2]  # print from start and till index 2 included
```

```python
df.loc[46]   # goto index 46 and print its value
```

<h2> using loc we are accessing element via index number</h2>
<h2> using iloc we are accessing element via row number</h2>

```python
df.iloc[:2]
```

```python
df.iloc[3]  # here 3 is row number. row number starts from 0.
```

```python
df.iloc[45]  # throw an error bcz there is not existing 45 row number
```

```python
df.loc[:2]  # till index 2
```

```python
df.iloc[:3]  # till row number 3 from row number =0
```

<h2>  .loc takes index and .iloc takes row number
